```python
import numpy as np
import matplotlib.pyplot as plt

def g_a(x):
    # p = 3
    return 1 + x - (1 / 9) * x ** 2

def fixed_point_iteration(g_func, x_0, num_iterations):
    x_seq = [x_0]
    for _ in range(num_iterations):
        x_seq.append(g_func(x_seq[-1]))
    x_seq = np.array(x_seq)
    return x_seq

# generate the sequence of fixed point iteration
N = 25
x_seq = fixed_point_iteration(g_a, 2, N)
p = 3

plt.figure(figsize=(7, 4))
plt.plot(x_seq, 'o-')
plt.xlabel('Iteration')
plt.ylabel('$x_n$')
plt.grid()
plt.show()
```

```python
# culculate and plot the errors
def calculate_errors(x, p):
    return np.abs(x - p)

error_seq = calculate_errors(x_seq, p)
plt.figure(figsize=(7, 4))
plt.plot(error_seq, 'o-')
plt.xlabel('Iteration')
plt.ylabel('$|x_n - p|$ log scale')
plt.yscale('log')
plt.grid()
plt.show()
```

```python
# calculate the convergence rate based on the last 10 iterations
it_seq = np.arange(1, len(x_seq) + 1)
coeffs = np.polyfit(it_seq[-10:], np.log(error_seq)[-10:], 1)
sigma = np.exp(coeffs[0])
print(f"Convergence rate: {sigma}")
```

```python
# check the convergence constant
[error_seq[k+1] / error_seq[k] for k in range(len(error_seq)-1)][-10:]
```

```python
import numpy as np
import matplotlib.pyplot as plt

def g_hat(x):
    # p \approx 1.25, 2.75
    return (x**2 + 3.5) / 4


x_interval = np.linspace(1, 3, 101)
plt.figure(figsize=(7, 4))
plt.plot(x_interval, g_hat(x_interval), 'b', label='$y = \hat{g}(x)$')
plt.plot(x_interval, x_interval, 'r--', label='$y = x$')
plt.xlabel('x')
plt.legend()
plt.grid()
plt.show()

```

```python
def fixed_point_iteration(g_func, x_0, num_iterations):
    x_seq = [x_0]
    for _ in range(num_iterations):
        x_seq.append(g_func(x_seq[-1]))
    x_seq = np.array(x_seq)
    return x_seq

# generate the sequence of fixed point iteration
x_seq = fixed_point_iteration(g_hat, x_0=1.25, num_iterations=15)
print(f"x approx to {x_seq[-1]}")

plt.figure(figsize=(7, 4))
plt.plot(x_seq, 'o-')
plt.xlabel('Iteration')
plt.ylabel('$x_n$')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def g(x):
    # p = 1/3
    return 2 * x - 3 * x**2

def fixed_point_iteration(g_func, x_0, num_iterations):
    x_seq = [x_0]
    for _ in range(num_iterations):
        x_seq.append(g_func(x_seq[-1]))
    x_seq = np.array(x_seq)
    return x_seq

# generate the sequence of fixed point iteration
N = 10
x_seq = fixed_point_iteration(g, 0.5, N)
p = 1 / 3

plt.figure(figsize=(7, 4))
plt.plot(x_seq, 'o-')
plt.xlabel('Iteration')
plt.ylabel('$x_n$')
plt.grid()
plt.show()
```

```python
# culculate and plot the errors
def calculate_errors(x, p):
    return np.abs(x - p)

error_seq = calculate_errors(x_seq, p)
plt.figure(figsize=(7, 4))
plt.plot(error_seq, 'o-')
plt.xlabel('Iteration')
plt.ylabel('$|x_n - p|$ log scale')
plt.yscale('log')
plt.grid()
plt.show()
```
