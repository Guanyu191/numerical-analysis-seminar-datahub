```python
i = np.arange(0, 101, 1)
n = 10**(1+0.03*i)

float_counts = 3*(n-1) + 5*n*(n-1)/2 + n*(n-1)*(2*n-1)/3
ratios = float_counts/(2*n**3/3)

plt.scatter(np.log(n), ratios)
plt.plot(np.log(n), n/n, c='green')
plt.xlabel("$log(n)$")
plt.ylabel("$ratio$")
plt.show()
```
