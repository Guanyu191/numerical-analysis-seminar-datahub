```python
import numpy as np
import matplotlib.pyplot as plt

# (a) Plot unit "circle" for ∞-norm
x = np.linspace(-1, 1, 100)
y1 = np.ones_like(x)
y2 = -np.ones_like(x)

plt.figure(figsize=(5, 5))
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'r')
plt.plot(y1, x, 'r')
plt.plot(y2, x, 'r')
plt.title('Unit "circle" in infty-norm')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# (b) Plot unit "circle" for 1-norm
x = np.linspace(-1, 1, 100)
y1 = 1 - np.abs(x)
y2 = - (1 - np.abs(x))

plt.figure(figsize=(5, 5))
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'r')
plt.title('Unit "circle" in 1-norm')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.grid()
plt.tight_layout()
plt.show()
```
